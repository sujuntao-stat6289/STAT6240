```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

### apyori library

```python
records=[['Beef', 'Chicken', 'Milk'],
         ['Beef', 'Cheese'],
         ['Cheese', 'Boots'],
         ['Beef', 'Chicken', 'Cheese'],
         ['Beef', 'Chicken', 'Clothes', 'Cheese', 'Milk'],
         ['Chicken', 'Clothes', 'Milk'],
         ['Chicken', 'Milk', 'Clothes']]
association_rules = apriori(records, min_support=0.3, min_confidence=0.8, min_lift=1, min_length=2)
association_results = list(association_rules)
for item in association_results:
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

```python
data=[['Beef', 'Chicken', 'Milk'],
      ['Beef', 'Cheese'],
      ['Cheese', 'Boots'],
      ['Beef', 'Chicken', 'Cheese'],
      ['Beef', 'Chicken', 'Clothes', 'Cheese', 'Milk'],
      ['Chicken', 'Clothes', 'Milk'],
      ['Chicken', 'Milk', 'Clothes']]
te = TransactionEncoder()
X =te.fit_transform(data)
colmns = te.columns_
df = pd.DataFrame(X,columns=colmns)
df.astype(np.uint8)
df
```

```python
result =apriori(df,min_support=0.3,use_colnames=True)
result
```

```python
association_rules(result,min_threshold=0.8)
```
